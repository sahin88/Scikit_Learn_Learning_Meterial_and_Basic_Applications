# 3 Regression Features and Labels - Practical Machine Learning Tutorial with Python

```python
import pandas 
import quandl
import sklearn
import math

df = quandl.get("WIKI/KO",trim_start ="2000-12-12", trim_end = "2014-12-30", authtoken='fzvgru14QQMetF9SeL_z')

df.head()


df=df[['Adj. Open','Adj. High','Adj. Low','Adj. Close','Adj. Volume']]
df['HL_PTC']=(df['Adj. High']-df['Adj. Close'])/df['Adj. Close']*100
df['PC_Percentage']=(df['Adj. Close']-df['Adj. Open'])/df['Adj. Open']*100
df=df[['Adj. Close','HL_PTC','PC_Percentage','Adj. Volume']]
df.head()
len(df)


forcast_col='Adj. Close'
df.fillna(-99999.9999,inplace=True)
forcast_out=int(math.ceil(0.1*len(df)))
df['Label']= df[forcast_col].shift(-forcast_out)
df.head()
```
